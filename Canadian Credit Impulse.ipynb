```python
data = pd.read_csv("Credit_Impulse.csv")
```

```python
data = data.iloc[::-1]
data
```

# Both Plots

```python
sns.set_style('whitegrid')
fig, ax = plt.subplots(figsize=(15,7))

#mdates.MonthLocator(interval=14)
#ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))
#ax.xaxis.set_major_locator(days)


plt.plot(data['Date'], data['Credit Impulse (% MoM)'], color='r')
plt.plot(data['Date'], data['Credit Impulse (% YoY)'], color='g')
plt.title("Canadian Credit Impulse MoM% (Red) and YoY% (Green)")
plt.xlabel('Date Range')
plt.ylabel('Credit Impulse % (Credit Issued / GDP)')

fig.autofmt_xdate()
n=6
ax.set_xticks(ax.get_xticks()[::n])



plt.axhline(0, color='k',linewidth=0.5)


plt.show()
```
